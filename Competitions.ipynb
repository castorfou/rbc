# Fichier adherents + classement

```python
import pandas as pd
adherents_filename = 'xls633be4e053e80.xlsx'
```

```python
pd.read_excel(adherents_filename).head()
```

```python
from rbc import add_classements
```

```python
adherents = add_classements(pd.read_excel(adherents_filename))
adherents.head()
```

```python
adherents.Catégorie.unique()
```

```python
adherents_adultes = adherents[~adherents.Catégorie.isin(['Junior 1', 'Minime 1', 'Minibad', 'Benjamin 2', 'Cadet 1', 'Junior 2', 'Benjamin 1', 'Poussin 1',    'Cadet 2', 'Minime 2'])]
```

```python
adherents_adultes.head()
```

```python
print(f"Adherents adulte: {adherents_adultes.shape[0]}")
```

# Merge des fichiers

```python
adherents_adultes = adherents_adultes.copy()
adherents_adultes["Email de contact"] = adherents_adultes["Email de contact"].str.lower()
competiteurs["Adresse e-mail"] = competiteurs["Adresse e-mail"].str.lower()

result = pd.merge(adherents_adultes, competiteurs, how="right", left_on="Email de contact", right_on="Adresse e-mail")
```

```python
print(f"Nombre adherents : {adherents_adultes.shape[0]}, Nombre competiteurs : {competiteurs.shape[0]}, Nombre liaison : {result.shape[0]}")
print(f"Pourcentage de liaison: {100*result.shape[0]/competiteurs.shape[0]:0.02f}")
```
